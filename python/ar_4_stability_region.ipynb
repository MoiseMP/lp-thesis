```python
import numpy as np
import itertools
```

```python
phi_values = np.linspace(-1,1,100)
```

```python
phi_grid = np.meshgrid(phi_values, phi_values, phi_values, phi_values)
```

```python
convergence_region = []
counter = 0

for phi1, phi2, phi3, phi4 in itertools.product(phi_values, repeat=4):
    if counter % 100 == 0:
        print(counter)
    
    roots = np.roots([-phi4, -phi3, -phi2, -phi1, 1])
    sizes = np.abs(roots)
    
    if np.all(np.abs(roots) > 1):
        convergence_region.append(sizes)
        
    counter += 1
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

a_vals = np.linspace(-1, 1, 400)
b_vals = np.linspace(-1, 1, 400)

A, B = np.meshgrid(a_vals, b_vals)
mask = (np.abs(B) <= np.abs(A))

plt.figure(figsize=(6,6))
plt.scatter(A[mask], B[mask], s=1)
plt.xlabel('a')
plt.ylabel('b')
plt.title('Region where all roots |x| ≤ 1')
plt.show()
```
